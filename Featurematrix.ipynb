```python
import os
import nltk
from nltk.corpus import *
from nltk.stem import PorterStemmer
nltk.download('stopwords')
```

```python
def pp(path):
  file=os.listdir(path)
  content=[]
  for i in range(len(file)):
    f=open(path+file[i])
    content.append(f.read())
    f.close()
  stopword=nltk.corpus.stopwords.words("english")
  semf=[]
  for i in content:
    semf.append([word.lower() for word in i.split() if word.isalpha() and word not in stopword])
  stemm=PorterStemmer()
  final=[]
  for i in semf:
    final.append(" ".join([stemm.stem(word) for word in i]))
  return final
```

```python
path="/content/drive/MyDrive/textdata/"
final=pp(path)
final
```

```python
#1) binary tdm
file=os.listdir(path)
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
cv= CountVectorizer(binary=True)
x=cv.fit_transform(final)
binary_tdm=pd.DataFrame(x.toarray().T,index=cv.get_feature_names_out(),columns=file)
binary_tdm
```

```python
#2) tdm
from  sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer()
x=cv.fit_transform(final)
tdm=pd.DataFrame(x.toarray().T,index=cv.get_feature_names_out(),columns=file)
tdm
```

```python
#3) tdm with Lnorm
from sklearn.feature_extraction.text import TfidfVectorizer
tf=TfidfVectorizer(use_idf=False,norm='l1')
x=tf.fit_transform(final)
tdm_l1=pd.DataFrame(x.toarray().T,index=tf.get_feature_names_out(),columns=file)
tdm_l1
```

```python
#4) tdm-maxNorm
tdm_maxnorm=tdm.iloc[:,:].div(tdm.max(axis=0),axis=1)
tdm_maxnorm
```

```python
#5) tdm log norm
import numpy as np
tdm_log=np.log10(1+tdm)
tdm_log
```

```python
#tf-idf matrix
from sklearn.feature_extraction.text import TfidfVectorizer
tf=TfidfVectorizer()
x=tf.fit_transform(final)
tf_idf=pd.DataFrame(x.toarray().T,index=tf.get_feature_names_out(),columns=file)
tf_idf
```

```python
#co-occ matrix
co_occ=np.dot(binary_tdm,binary_tdm.T)
np.fill_diagonal(co_occ,0)
co_occ=pd.DataFrame(co_occ,index=binary_tdm.index,columns=binary_tdm.index)
co_occ
```
